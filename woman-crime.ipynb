# Importing required libraries

```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas
import matplotlib

%matplotlib inline
%pylab inline

matplotlib.style.use('ggplot')
```

```python
woman_data=pandas.read_csv('woman.csv')
```

```python
woman_data.head()
```

```python
state_wise=woman_data[woman_data['DISTRICT']=='TOTAL']
```

```python
state_wise.head()
```

```python
state_wise_ap=state_wise[state_wise['STATE/UT']=='ANDHRA PRADESH']
state_wise_ap.head()
```

```python
plt.scatter(state_wise_ap['Year'],state_wise_ap['Rape'])
plt.title("Rape vs Year in Andhra Pradesh")
plt.ylabel("Rape")
plt.xlabel("Year")
plt.rcParams["figure.figsize"]=100,100

```
